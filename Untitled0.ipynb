```python
x=df.drop("Outcome",axis=1)
y=df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
from sklearn.linear_model import LogisticRegression
leo=LogisticRegression()
leo.fit(x_train,y_train)
y_prd=leo.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_prd)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
erripuk=DecisionTreeClassifier()
erripuk.fit(x_train,y_train)
y_pred=erripuk.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_preed=xgb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_preed)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_preedd=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_preedd)*100)
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(x_train,y_train)
y_preddd=svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_preddd)*100)
```
